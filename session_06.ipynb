```python
import pandas as pd
df = pd.read_csv('./DataSet/real_estate_data.csv',sep=',')
df.head()
```

```python
df.info()
```

```python
list(df.columns)
map_dict ={
    'last_price': '最终价格',
    'total_area': '总面积',
    'first_day_exposition': '首次挂牌日期',
    'rooms': '房间数',
    'ceiling_height': '层高',
    'floors_total': '总楼层数',
    'living_area': '居住面积',
    'floor': '所在楼层',
    'is_apartment': '是否为公寓',
    'studio': '是否为单间',
    'open_plan': '是否为开放式布局',
    'kitchen_area': '厨房面积',
    'balcony': '是否有阳台',
    'locality_name': '所在地区名称',
    'days_exposition': '挂牌天数'
}
df.rename(columns=map_dict,inplace=True)
```

```python
df.info()
```

```python
df.是否为公寓.unique()
```

```python
df.首次挂牌日期.unique()
df['首次挂牌日期'] = pd.to_datetime(df.首次挂牌日期,errors='coerce')
df.info()
```

```python
df.首次挂牌日期.unique()
```

```python
# 针对总面积的平均值来计算是否属于公寓
median_total_price = df.最终价格.median()
median_total_area = df.总面积.median()
df['是否为公寓'] = df['是否为公寓'].fillna(
    (df.最终价格 > median_total_price)&(df.总面积 > median_total_area)
).astype('bool')
```
