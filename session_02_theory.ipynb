# List and List Operation

```python
# List example
x = [1, 2, 3, 4, 5]
print(len(x),x)
print(x[0])
print(x[1:3])
```

```python
# append() 方法apped 是在列表末尾添加一个元素
x = [1, 2, 3, 4, 5]
x.append(600)
x
```

```python
# insert()函数  在列表指定位置的前面插入一个元素 ist.insert(index, element)
x = [1,2,3]
x.insert(2,5) #parameter first 1 is index second is element  for exampple
x
```

```python
# extend() 函数  在列表末尾一次性追加另一个序列中的多个值
x=[1,2,3]
x.extend([4,5,6])
x
```

```python
# remove() 函数  删除列表中某个值的第一个匹配项
x = [1,2,3,4,5,6,7,8,9,5]
x.remove(5)
x
```

```python
 # sort() 函数 对列表进行排序
x = [5,4,3,2,1]
x.sort()  # 不带参数 默认情况升序

```

```python
# reverse() 函数  列表元素进行反向排序
x = [1,2,3,5,9]
x.reverse()
x
```

```python
# clear() 函数  删除列表中的所有元素
x = [1,2,3,4,5]
x.clear()
x
```

# dictionary 字典

```python
# 创建字典
people = {'name': 'Bob', 'age': 20,'sex': 'Male', 'city': 'New York'}
print(people['name']) # 打印出值
people['name'] = 'Zane'
print(people)
print(people["name"])

for item in people:
    print(item,":",people[item])
print("===================================================")
for key, value in people.items():
    print(key,":",value)
```
